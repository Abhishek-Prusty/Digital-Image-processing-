# Question - 1

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### Original sky.jpeg image

```python
img1=cv2.imread('../input_data/sky.jpeg',0)
plt.imshow(img1,cmap='gray',vmin=0,vmax=255)
plt.axis('off')
```

###           Original sky2.jpg image

```python
img2=cv2.imread('../input_data/sky2.jpg',0)
plt.imshow(img2,cmap='gray',vmin=0,vmax=255)
plt.axis('off')
```

#### 1.a)

```python
f1=np.array([[1,2,1],[2,4,2],[1,2,1]])/16.0
modifiedf1_1=np.array([[2,2,2],[4,4,4],[2,2,2]])/16.0
modifiedf1_2=np.array([[1,2,1],[2,4,2],[-1,-2,-1]])/16.0

outim = cv2.filter2D(img1,-1,f1)
outim0 = cv2.filter2D(img2,-1,f1)
outim1 = cv2.filter2D(img1,-1,modifiedf1_1)
outim2 = cv2.filter2D(img1,-1,modifiedf1_2)
outim3 = cv2.filter2D(img2,-1,modifiedf1_1)
outim4 = cv2.filter2D(img2,-1,modifiedf1_2)

f, axs = plt.subplots(3,2,figsize=(15,15))

axs[0,0].axis('off')
axs[0,0].set_title('f1 applied to sky.jpg')
axs[0,0].imshow(outim,cmap='gray',vmin=0,vmax=255)

axs[0,1].axis('off')
axs[0,1].set_title('f1 applied to sky2.jpg')
axs[0,1].imshow(outim0,cmap='gray',vmin=0,vmax=255)

axs[1,0].axis('off')
axs[1,0].set_title('Modified filter1_1 applied to sky.jpg')
axs[1,0].imshow(outim1,cmap='gray',vmin=0,vmax=255)

axs[1,1].set_title('Modified filter1_1 applied to sky2.jpg')
axs[1,1].axis('off')
axs[1,1].imshow(outim3,cmap='gray',vmin=0,vmax=255)

axs[2,0].set_title('Modified filter1_2 applied to sky.jpg')
axs[2,0].axis('off')
axs[2,0].imshow(outim2,cmap='gray',vmin=0,vmax=255)

axs[2,1].set_title('Modified filter1_2 applied to sky2.jpg')
axs[2,1].axis('off')
axs[2,1].imshow(outim4,cmap='gray',vmin=0,vmax=255)
plt.tight_layout()
```

```python
f2=np.array([[1,4,6,4,1],[4,16,24,16,4],[6,24,36,24,6],[4,16,24,16,4],[1,4,6,4,1]])/256.0
modifiedf2_1=np.array([[1,4,6,4,1],[4,16,24,16,4],[6,24,36,24,6],[4,16,24,16,4],[1,4,6,4,1]])/128.0
modifiedf2_2=np.array([[1,4,8,4,1],[4,16,32,16,4],[6,24,48,24,6],[4,16,32,16,4],[1,4,8,4,1]])/256.0

outim = cv2.filter2D(img1,-1,f2)
outim0 = cv2.filter2D(img2,-1,f2)
outim1 = cv2.filter2D(img1,-1,modifiedf2_1)
outim2 = cv2.filter2D(img1,-1,modifiedf2_2)
outim3 = cv2.filter2D(img2,-1,modifiedf2_1)
outim4 = cv2.filter2D(img2,-1,modifiedf2_2)

f, axs = plt.subplots(3,2,figsize=(15,15))

axs[0,0].axis('off')
axs[0,0].set_title('f2 applied to sky.jpg')
axs[0,0].imshow(outim,cmap='gray',vmin=0,vmax=255)

axs[0,1].axis('off')
axs[0,1].set_title('f2 applied to sky2.jpg')
axs[0,1].imshow(outim0,cmap='gray',vmin=0,vmax=255)

axs[1,0].axis('off')
axs[1,0].set_title('Modified filter2_1 applied to sky.jpg')
axs[1,0].imshow(outim1,cmap='gray',vmin=0,vmax=255)

axs[1,1].set_title('Modified filter2_1 applied to sky2.jpg')
axs[1,1].axis('off')
axs[1,1].imshow(outim3,cmap='gray',vmin=0,vmax=255)

axs[2,0].set_title('Modified filter2_2 applied to sky.jpg')
axs[2,0].axis('off')
axs[2,0].imshow(outim2,cmap='gray',vmin=0,vmax=255)

axs[2,1].set_title('Modified filter2_2 applied to sky2.jpg')
axs[2,1].axis('off')
axs[2,1].imshow(outim4,cmap='gray',vmin=0,vmax=255)
plt.tight_layout()
```

```python
f3=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])/8.0
modifiedf3_1=np.array([[-1,0,2],[-3,0,2],[-1,0,2]])/8.0
modifiedf3_2=np.array([[-4,1,4],[-4,1,4],[-4,1,4]])/8.0

outim = cv2.filter2D(img1,-1,f3)
outim0 = cv2.filter2D(img2,-1,f3)
outim1 = cv2.filter2D(img1,-1,modifiedf3_1)
outim2 = cv2.filter2D(img1,-1,modifiedf3_2)
outim3 = cv2.filter2D(img2,-1,modifiedf3_1)
outim4 = cv2.filter2D(img2,-1,modifiedf3_2)

f, axs = plt.subplots(3,2,figsize=(15,15))

axs[0,0].axis('off')
axs[0,0].set_title('f3 applied to sky.jpg')
axs[0,0].imshow(outim,cmap='gray',vmin=0,vmax=255)

axs[0,1].axis('off')
axs[0,1].set_title('f3 applied to sky2.jpg')
axs[0,1].imshow(outim0,cmap='gray',vmin=0,vmax=255)

axs[1,0].axis('off')
axs[1,0].set_title('Modified filter3_1 applied to sky.jpg')
axs[1,0].imshow(outim1,cmap='gray',vmin=0,vmax=255)

axs[1,1].set_title('Modified filter3_1 applied to sky2.jpg')
axs[1,1].axis('off')
axs[1,1].imshow(outim3,cmap='gray',vmin=0,vmax=255)

axs[2,0].set_title('Modified filter3_2 applied to sky.jpg')
axs[2,0].axis('off')
axs[2,0].imshow(outim2,cmap='gray',vmin=0,vmax=255)

axs[2,1].set_title('Modified filter3_2 applied to sky2.jpg')
axs[2,1].axis('off')
axs[2,1].imshow(outim4,cmap='gray',vmin=0,vmax=255)
plt.tight_layout()
```

*Answer*
<br>
 - Some filters sum to 1 because if they don't ,they would overflow at some pixel values.
<br>
 - Some filters sum to 0 because most of them are used for edge detection or for finding change in the intensity of pixels in a region. If they dont sum to 0 they would give a weighted difference which will lead to wrong output.
