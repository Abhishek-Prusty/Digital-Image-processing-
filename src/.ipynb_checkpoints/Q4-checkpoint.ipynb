```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#function that detects the edge
def detect_edge(img,filt,save_name):
    win=filt.shape[0]
    final_img=np.zeros(img.shape)
    wind_size=win
    new_img = cv2.copyMakeBorder(img,wind_size//2,wind_size//2,wind_size//2,wind_size//2,cv2.BORDER_REPLICATE)
    
    print(new_img.shape)
    for i in range(win//2,len(new_img)-win//2):
        for j in range(win//2,len(new_img[i])-win//2):
            if win%2==0:
                neighbourhood=new_img[i-win//2:i+win//2,j-win//2:j+win//2]
            else:
                neighbourhood=new_img[i-win//2:i+1+win//2,j-win//2:j+1+win//2]
            count=0
            for nei0 in range(neighbourhood.shape[0]):
                for nei1 in range(neighbourhood.shape[1]):
                    count+=neighbourhood[nei0][nei1]*filt[nei0][nei1]
            final_img[i-win//2][j-win//2]=count
    
    #plt.imshow(np.array(final_img),cmap='gray',vmin=0, vmax=255)
    cv2.imwrite(save_name,final_img)
```

```python
#image samples
im1=cv2.imread('../input_data/barbara.jpg',0)
#original image 1
plt.axis('off')
plt.imshow(im1,cmap='gray')
```

```python
#filters
px=np.array([[-1,0,1] for i in range(3)])
py=np.array([[1,1,1],[0,0,0],[-1,-1,-1]]) 
sx=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sy=np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
lp=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
lp1=np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
rx=np.array([[0,1],[-1,0]])
ry=np.array([[1,0],[0,-1]])
```

```python
detect_edge(im1,px,'../output/prewitt_x.png')
detect_edge(im1,py,'../output/prewitt_y.png')
detect_edge(im1,sx,'../output/sobel_x.png')
detect_edge(im1,sy,'../output/sobel_y.png')
detect_edge(im1,lp,'../output/laplacian.png')
detect_edge(im1,lp1,'../output/laplacian_with_diag.png')
detect_edge(im1,rx,'../output/roberts_x.png')
detect_edge(im1,ry,'../output/roberts_y.png')
```

```python
im1=cv2.imread('../output/prewitt_x.png')
im2=cv2.imread('../output/prewitt_y.png')
im3=cv2.imread('../output/sobel_x.png')
im4=cv2.imread('../output/sobel_y.png')
im5=cv2.imread('../output/laplacian.png')
im6=cv2.imread('../output/laplacian_with_diag.png')
im7=cv2.imread('../output/roberts_x.png')
im8=cv2.imread('../output/roberts_y.png')
```

The Laplcaian edge detector is not doing very well because, it is detecting edges in all directions, thereby reducing the intensity of edges along x-axis.
