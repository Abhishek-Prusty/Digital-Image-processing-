```python
import time
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### Function to apply mean filter

```python
def conv2d(img,k):
    to=time.time()
    filt=np.array([[1.0/(k*k) for i in range(k)]for j in range(k)])
    win=k
    final_img=np.zeros(img.shape)
    wind_size=win
    new_img = cv2.copyMakeBorder(img,wind_size//2,wind_size//2,wind_size//2,wind_size//2,cv2.BORDER_REPLICATE)
    
    print(new_img.shape)
    for i in range(win//2,len(new_img)-win//2):
        for j in range(win//2,len(new_img[i])-win//2):
            if win%2==0:
                neighbourhood=new_img[i-win//2:i+win//2,j-win//2:j+win//2]
            else:
                neighbourhood=new_img[i-win//2:i+1+win//2,j-win//2:j+1+win//2]
            count=0
            for nei0 in range(neighbourhood.shape[0]):
                for nei1 in range(neighbourhood.shape[1]):
                    count+=neighbourhood[nei0][nei1]*filt[nei0][nei1]
            final_img[i-win//2][j-win//2]=count
    
    
    #return final_img
    return time.time()-to
```

### Calculate time taken by the above function

```python
im1=cv2.imread('../input_data/barbara.jpg',0)
t0=time.time()
conv2d(im1,10)
print(time.time()-t0)
```

### Plot

```python
im1=cv2.imread('../input_data/barbara.jpg',0)
im1=cv2.resize(im1,(256,256))
x=[5,10,20,40,80]
y=[]
for i in range(len(x)):
    print(x[i])
    ti=conv2d(im1,x[i])
    y.append(ti)
plt.plot(x,y)
plt.show()
```

### Efficient version of the above filter

```python
def conv2deff(img,k):
    to=time.time()
    filt=np.array([[1.0 for i in range(k)]for j in range(k)])
    win=k
    final_img=np.zeros(img.shape)
    wind_size=win
    new_img = cv2.copyMakeBorder(img,wind_size//2,wind_size//2,wind_size//2,wind_size//2,cv2.BORDER_REPLICATE)
    
    print(new_img.shape)
    for i in range(win//2,len(new_img)-win//2):
        #print(i)
        count=0
        for j in range(win//2,len(new_img[i])-win//2):
            
            if win%2==0:
                if j==win//2:
                    neighbourhood=new_img[i-win//2:i+win//2,j-win//2:j+win//2]
                    count=mult(neighbourhood,filt)
                    final_img[i-win//2][j-win//2]=count/(1.0*k*k)
                else:
                    
                    count=count+np.sum(new_img[i-win//2:i+win//2,j+win//2 -1 :j+win//2])
                    count-=np.sum(new_img[i-win//2:i+win//2,j-win//2 -1 :j-win//2])
                    final_img[i-win//2][j-win//2]=count/(1.0*k*k)
            else:
                if j==win//2:
                    neighbourhood=new_img[i-win//2:i+win//2,j-win//2:j+win//2]
                    count=mult(neighbourhood,filt)
                    final_img[i-win//2][j-win//2]=count/(1.0*k*k)
                else:
                    
                    count=count+np.sum(new_img[i-win//2:i+win//2,j+win//2 -1 :j+win//2])
                    count-=np.sum(new_img[i-win//2:i+win//2,j-win//2 -1 :j-win//2])
                    final_img[i-win//2][j-win//2]=count/(1.0*k*k)
                    
    
    
    #plt.imshow(np.array(final_img),cmap='gray',vmin=0, vmax=255)
    #plt.show()
    #return final_img
    return time.time()-to

def mult(m1,m2):
    out=0
    for i in range(len(m1)):
        for j in range(len(m1[i])):
            out+=m1[i][j]*m2[i][j]
    #print(out)
    return out
```

### Median filter

```python
def unoptimized_median(img,win):
    to=time.time()
    #filt=np.array([[1.0/(k*k) for i in range(k)]for j in range(k)])
    
    final_img=np.zeros(img.shape)
    wind_size=win
    new_img = cv2.copyMakeBorder(img,wind_size//2,wind_size//2,wind_size//2,wind_size//2,cv2.BORDER_REPLICATE)
    for i in range(win//2,len(new_img)-win//2):
        #print(i)
        for j in range(win//2,len(new_img[i])-win//2):
            if win%2==0:
                neighbourhood=new_img[i-win//2:i+win//2,j-win//2:j+win//2]
            else:
                neighbourhood=new_img[i-win//2:i+1+win//2,j-win//2:j+1+win//2]
            count=median(neighbourhood)
            final_img[i-win//2][j-win//2]=count
    
    #plt.imshow(np.array(final_img),cmap='gray',vmin=0, vmax=255)
    #plt.show()
    #return final_img
    return time.time()-to

def median(arr):
    narr=[]
    for i in arr:
        for j in i:
            narr.append(j)
    narr.sort()
    return narr[int(len(narr)/2)]
```

```python
im1=cv2.imread('../input_data/barbara.jpg',0)
im1=cv2.resize(im1,(200,200))
unoptmx=[3,5,7,21,31,55]
unoptmy=[]
for i in unoptmx:
    print(i)
    t1=unoptimized_median(im1,i)
    unoptmy.append(t1)
```

### Optimised version of the above median filter

```python
def optimized_median(img,k):
    to=time.time()
    #filt=np.array([[1.0/(k*k) for i in range(k)]for j in range(k)])
    
    final_img=np.zeros(img.shape)
    win=k
    wind_size=win
    new_img = cv2.copyMakeBorder(img,wind_size//2,wind_size//2,wind_size//2,wind_size//2,cv2.BORDER_REPLICATE)
    for i in range(win//2,len(new_img)-win//2):
        medarray=[]
        for j in range(win//2,len(new_img[i])-win//2):
           
            if j==win//2:
                #neighbourhood=new_img[i-win//2:i+1+win//2,j-win//2:j+1+win//2]
                for it1 in range(j-win//2,j+1+win//2):
                    for it2 in range(i-win//2,i+1+win//2):
                        medarray.append(new_img[it2][it1])
            else:
                medarray=medarray[k:]
                medarray.extend(new_img[i-win//2:i+win//2,j+win//2:j+1+win//2])
            
            compute_array=list(medarray)
            compute_array.sort()
            final_img[i-win//2][j-win//2]=compute_array[int(len(compute_array)/2)]
    
    #return final_img
    return time.time()-to
```

### Plot

```python
im1=cv2.imread('../input_data/barbara.jpg',0)
im1=cv2.resize(im1,(200,200))
optmx=[3,5,7,21,31,55]
optmy=[]
for i in optmx:
    print(i)
    t1=optimized_median(im1,i)
    optmy.append(t1)
plt.figure(figsize=(11,11))
plt.plot(optmx,optmy)
plt.plot(unoptmx,unoptmy)
plt.legend(['Optimized Median', 'Unoptimized Median'], loc='upper left')
```
